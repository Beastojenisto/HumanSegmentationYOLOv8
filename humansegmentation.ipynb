# **Creating The configuration file for the YOLOv8 model**

```python
import yaml

data_yaml = dict(
    train ='/kaggle/working/Pedestrian-Detection-5-6/train',
    val ='/kaggle/working/Pedestrian-Detection-5-6/valid',
    test='/kaggle/working/Pedestrian-Detection-5-6/valid',
    names =  ['adult','child'],
    nc = 2,
)

with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
    
%cat data.yaml
```

# **Loading the YOLOv8 segmentation model and training it for 50 epochs**

```python
from ultralytics import YOLO
model = YOLO('yolov8m-seg.pt')
model.train(data='data.yaml', epochs=50, imgsz=640, batch=16,exist_ok=True)
```
